```python
import numpy as np
import matplotlib.pyplot as plt
from itertools import product
from sklearn.linear_model import Perceptron
```

```python
# AND gate
data_and = [[0,0],[0,1],[1,0],[1,1]]
labels_and = [0,0,0,1]

plt.scatter([p[0] for p in data_and], [p[1] for p in data_and],
            c=labels_and, cmap='viridis', s=100)
plt.title("AND Gate Perceptron")
plt.xlabel("Input 1"); plt.ylabel("Input 2")
plt.show()

clf_and = Perceptron(max_iter=40, random_state=22)
clf_and.fit(data_and, labels_and)
print("Learning Score AND:", clf_and.score(data_and, labels_and))
```

```python
# OR gate
data_or = [[0,0],[0,1],[1,0],[1,1]]
labels_or = [0,1,1,1]

plt.scatter([p[0] for p in data_or], [p[1] for p in data_or],
            c=labels_or, cmap='viridis', s=100)
plt.title("OR Gate Perceptron")
plt.xlabel("Input 1"); plt.ylabel("Input 2")
plt.show()

clf_or = Perceptron(max_iter=40, random_state=22)
clf_or.fit(data_or, labels_or)
print("Learning Score OR:", clf_or.score(data_or, labels_or))
```

```python
# XOR gate
data_xor = [[0,0],[0,1],[1,0],[1,1]]
labels_xor = [0,1,1,0]

plt.scatter([p[0] for p in data_xor], [p[1] for p in data_xor],
            c=labels_xor, cmap='viridis', s=100)
plt.title("XOR Gate Perceptron")
plt.xlabel("Input 1"); plt.ylabel("Input 2")
plt.show()

clf_xor = Perceptron(max_iter=40, random_state=22)
clf_xor.fit(data_xor, labels_xor)
print("Learning Score XOR:", clf_xor.score(data_xor, labels_xor))
```

```python
# Decision boundary heatmap function
def plot_heatmap(classifier, title):
    x_values = np.linspace(0, 1, 100)
    y_values = np.linspace(0, 1, 100)
    point_grid = list(product(x_values, y_values))
    distances = classifier.decision_function(point_grid)
    abs_distances = [abs(d) for d in distances]
    distances_matrix = np.reshape(abs_distances, (100, 100))
    heatmap = plt.pcolormesh(x_values, y_values, distances_matrix, shading='auto')
    plt.colorbar(heatmap)
    plt.title(title)
    plt.xlabel("Input 1"); plt.ylabel("Input 2")
    plt.show()
```

```python
# Heatmaps
print("AND decision function:", clf_and.decision_function([[0,0],[1,1],[0.5,0.5]]))
plot_heatmap(clf_and, "AND Gate Decision Boundary")

print("OR decision function:", clf_or.decision_function([[0,0],[1,1],[0.5,0.5]]))
plot_heatmap(clf_or, "OR Gate Decision Boundary")

print("XOR decision function:", clf_xor.decision_function([[0,0],[1,1],[0.5,0.5]]))
plot_heatmap(clf_xor, "XOR Gate Decision Boundary")
```
