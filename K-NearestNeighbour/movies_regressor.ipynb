```python
from movies import movie_dataset, movie_ratings
import numpy as np
from sklearn.neighbors import KNeighborsRegressor
```

```python
# Euclidean distance function
def distance(movie1, movie2):
    squared_difference = 0
    for i in range(len(movie1)):
        squared_difference += (movie1[i] - movie2[i]) ** 2
    return squared_difference ** 0.5
```

```python
# Weighted KNN regression prediction
def predict(unknown, dataset, movie_ratings, k):
    distances = []
    # Loop through all points in the dataset
    for title, movie in dataset.items():
        distance_to_point = distance(movie, unknown)
        distances.append([distance_to_point, title])
            # Sort by distance (ascending)
    distances.sort(key=lambda x: x[0])
    
    # Take only the k closest points
    neighbors = distances[:k]
    
    numerator = 0
    denominator = 0
    for neighbor in neighbors:
        title = neighbor[1]
        rating = movie_ratings[title]
        distance_to_neighbor = neighbor[0]
        
        # Avoid division by zero if distance is 0
        if distance_to_neighbor == 0:
            return rating
        
        numerator += rating / distance_to_neighbor
        denominator += 1 / distance_to_neighbor
    
    return numerator / denominator
```

```python
# Example usage
print(predict([0.016, 0.300, 1.022], movie_dataset, movie_ratings, 5))
```

```python
# Convert your dataset and ratings into NumPy arrays
# sklearn expects arrays, not dicts
X = np.array(list(movie_dataset.values()))   
y = np.array(list(movie_ratings.values()))  

# Create KNN regressor (k=5, distance-weighted)
regressor = KNeighborsRegressor(n_neighbors = 5, weights="distance")

# Fit model on dataset
regressor.fit(X, y)

# Predict ratings for new movies 
predictions = regressor.predict([
    [0.016, 0.300, 1.022],
    [0.0004092981, 0.283, 1.0112],
    [0.00687649, 0.235, 1.0112]
])
print(predictions)
```
