```python
import numpy as np
import pandas as pd
np.random.seed(10) #Yeah hr baar naya data set generate karne se rokhta hai! 
n = 100 # Yeah number of rows hai!
independent_variable = np.random.uniform(20, 40, n)
dependent_varaible = 90 - (independent_variable * 2) + np.random.normal(0, 5, n)
#upr formuale use kiya hai taki inme relation bnna rahe taki graph random na ho!
df = pd.DataFrame({
    "Independent_C": independent_variable,
    "Dependent_C": dependent_varaible
}   )
df.to_excel("lab_data_set.xlsx", index=False)
print(df.head(10))
print(df.describe())
```

```python
import scipy.stats as stats
mean_independent = np.mean(independent_variable)
median_independet = np.median(independent_variable)
mode_independent = stats.mode(independent_variable)
std_independent = np.std(independent_variable)
var_independent = np.var(independent_variable)
print(np.mean(independent_variable))
print(f"Median of Independent Variable: {median_independet}")
print(f"Mode of Independent Variable: {mode_independent}")
print(f"Standard Deviation of Independent Variable: {std_independent}")
print(f"Variance of Independent Variable: {var_independent}")
```

```python
import matplotlib.pyplot as plot
plot.scatter(independent_variable, dependent_varaible)
plot.xlabel("Independent_C")
plot.ylabel("Dependent_C")
plot.title("Scatter Plot of Independent vs Dependent Variable")
plot.show()
```

```python
import matplotlib.pyplot as plt
slope, intercept, r_value, p_value, std_err = stats.linregress(independent_variable, dependent_varaible)
print(f"Slope: {slope}, Intercept: {intercept}")
regression_line = slope * independent_variable + intercept
plt.plot(independent_variable, regression_line, color='red', label='Regression Line')
```

```python
t_stat = slope / std_err
df = n - 2
print(f"T-statistic: {t_stat}")
print(f"Degrees of freedom: {df}")
t_p_value = 2 * (1 - stats.t.cdf(abs(t_stat), df))
print(f"Calculated P-value: {t_p_value}")
```
