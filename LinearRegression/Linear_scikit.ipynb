```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.linear_model import LinearRegression
```

```python
temperature = np.array(range(60, 100, 2)).reshape(-1, 1)
sales = [65, 58, 46, 45, 44, 42, 40, 40, 36, 38, 38, 28, 30, 22, 27, 25, 25, 20, 15, 5]
```

```python
model = LinearRegression()
model.fit(temperature, sales)
```

```python
sales_pred = model.predict(temperature)

# plot
plt.scatter(temperature, sales, label="Data")
plt.plot(temperature, sales_pred, color="red", label="Regression Line")
plt.xlabel("Temperature")
plt.ylabel("Sales")
plt.legend()
plt.show()
```
