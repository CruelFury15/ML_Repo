## Naive Bayes classifier 
Itis a supervised machine learning algorithm that leverages Bayes’ Theorem to make predictions and classifications.

 This type of classifier could be extremely helpful for a company that is curious about the public reaction to a new product.

```python
from collections import Counter
```

```python
pos_counter = Counter({
    "amazing": 10,
    "great": 8,
    "love": 7,
    "crib": 5,
    "comfortable": 6,
    "happy": 4
})

neg_counter = Counter({
    "bad": 9,
    "terrible": 7,
    "hate": 6,
    "crib": 2,
    "uncomfortable": 5,
    "sad": 3
})
```

```python
review = "This crib was amazing"

#Priors
percent_pos = 0.5
percent_neg = 0.5

#Total word counts
total_pos = sum(pos_counter.values())
total_neg = sum(neg_counter.values())

#Vocabulary size (unique words across both counters)
vocab_size = len(set(pos_counter.keys()) | set(neg_counter.keys()))

#Initialize probabilities with priors
pos_probability = percent_pos
neg_probability = percent_neg
```

```python
#Tokenize review
review_words = review.lower().split()

#Naive Bayes 
for word in review_words:
    word_in_pos = pos_counter.get(word, 0)
    word_in_neg = neg_counter.get(word, 0)

    pos_probability *= (word_in_pos + 1) / (total_pos + vocab_size)
    neg_probability *= (word_in_neg + 1) / (total_neg + vocab_size)
```

```python
# Normalize to get final probabilities
total_prob = pos_probability + neg_probability
pos_probability /= total_prob
neg_probability /= total_prob

print(f"Positive probability: {pos_probability:.4f}")
print(f"Negative probability: {neg_probability:.4f}")

if pos_probability > neg_probability:
    print("Review classified as POSITIVE")
else:
    print("Review classified as NEGATIVE")
```

```python
#Using Scikit learn
from sklearn.feature_extraction.text import CountVectorizer

review = "This crib was amazing"

# Build a corpus of synthetic reviews by repeating each word
neg_docs = [word for word, count in neg_counter.items() for _ in range(count)]
pos_docs = [word for word, count in pos_counter.items() for _ in range(count)]
training_docs = neg_docs + pos_docs  # 32 neg + 40 pos = 72 total

counter = CountVectorizer()
counter.fit(training_docs)
print(counter.vocabulary_)
review_counts = counter.transform([review])
print(review_counts.toarray())
training_counts = counter.transform(training_docs)
```

```python
from sklearn.naive_bayes import MultinomialNB

review_counts = counter.transform([review])
classifier = MultinomialNB()
# Labels match training_docs: 0 = negative, 1 = positive
training_labels = [0] * len(neg_docs) + [1] * len(pos_docs)
classifier.fit(training_counts, training_labels)
print(classifier.predict(review_counts))
print(classifier.predict_proba(review_counts))
```
