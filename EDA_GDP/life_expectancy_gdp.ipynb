```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

We will analyze data on GDP and life expectancy from the World Health Organization and the World Bank to try and identify the relationship between the GDP and life expectancy of six countries.

```python
life_df = pd.read_csv("all_data.csv")
print(life_df.head())

df_columns = life_df.columns.tolist()
print("\n Columns Present in the Dataset: ", df_columns)

print(life_df.info())
```

```python
print(life_exp.describe(),"\n")
print(gdp.describe(),"\n")
print(country.unique(),"\n")

life_df['Country'] = life_df['Country'].replace("United States of America", "USA")
print("Checking USA column: \n",life_df[life_df['Country']== "USA"])
```

```python
life_exp = life_df['Life expectancy at birth (years)']
gdp = life_df['GDP']
country = life_df['Country']
year = life_df['Year']
```

```python
#Has life expectancy increased over time in the six nations? 
plt.figure(figsize=(10,6))
sns.scatterplot(
    data=life_df,
    x="GDP",
    y="Life expectancy at birth (years)",
    hue="Country",   
    palette="Set2",
    marker = "s"
)
plt.title("Life Expectancy Over Time (Six Nations)")
plt.xlabel("Year")
plt.ylabel("Life Expectancy (years)")
plt.legend(title="Country")
plt.show()
```

```python
#Has GDP increased over time in the six nations?
sns.lineplot(
    data=life_df,
    x="Year",
    y="GDP",
    hue="Country",
    marker="o"
)
plt.title("GDP Over Time in Six Nations")
plt.xlabel("Year")
plt.ylabel("GDP")
plt.legend(title="Country")
plt.show()
```

```python
#average GDP per country
sns.barplot(
    data=life_df,
    x="Country",
    y="GDP",
    estimator="mean",
    palette="muted",
    hue = "Country"
)  
#Using Log Scale  
plt.yscale("log") 
plt.title("Average GDP by Country")
plt.xlabel("Country")
plt.ylabel("Average GDP")
plt.show()
```

```python
# Average Life Expectancy by Country
avg_life_exp = life_df.groupby("Country")["Life expectancy at birth (years)"].mean()

print("Average Life Expectancy in Six Nations:\n", avg_life_exp)
import seaborn as sns
import matplotlib.pyplot as plt

sns.barplot(
    data=life_df,
    x="Country",
    y="Life expectancy at birth (years)",
    estimator="mean",
    palette="muted",
    hue="Country"
)
plt.title("Average Life Expectancy by Country")
plt.xlabel("Country")
plt.ylabel("Average Life Expectancy (years)")
plt.show()
```

```python
#What is the distribution of that life expectancy
import seaborn as sns
import matplotlib.pyplot as plt

sns.histplot(
    data=life_df,
    x="Life expectancy at birth (years)",
    bins=10,          
    kde=True,       
    color="skyblue"
)
plt.title("Distribution of Life Expectancy")
plt.xlabel("Life Expectancy (years)")
plt.ylabel("Frequency")
plt.show()
```

```python
#What is the distribution of that life expectancy using BoxPlot
sns.boxplot(
    data=life_df,
    x="Country",
    y="Life expectancy at birth (years)",
    palette="muted",
    hue="Country"
)
plt.title("Life Expectancy Distribution by Country")
plt.xlabel("Country")
plt.ylabel("Life Expectancy (years)")
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.violinplot(
    data=life_df,
    x="Country",
    y="Life expectancy at birth (years)",
    palette="muted",
    hue="Country"
)

plt.title("Life Expectancy Distribution by Country")
plt.xlabel("Country")
plt.ylabel("Life Expectancy (years)")
plt.xticks(rotation=45)
plt.show()
```
